```python
import requests
```

```python
response=requests.get("https://coronavirus.m.pipedream.net/")
```

```python
data=response.json()
```

```python
data
```

```python
#Country,Province,Confirmed,Deaths,Cast_Fatality_Ratio

with open("covid.csv", "w") as f:
    f.write("Country,Province,Confirmed,Deaths,Cast_Fatality_Ratio\n")
    for sample in data["rawData"]:
        Country = sample["Country_Region"]
        Province = sample["Province_State"]
        Confirmed = sample["Confirmed"]
        Deaths = sample["Deaths"]
        Cast_Fatality_Ratio = sample["Case_Fatality_Ratio"]
        
        f.write(f"{Country},{Province},{Confirmed},{Deaths},{Cast_Fatality_Ratio}\n")
```
