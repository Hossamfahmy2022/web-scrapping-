```python
# Get name from page title
title = soup.find('title').get_text()
title
```

```python
file_name = title.split(' | ')[0] + '.csv'
file_name
```

```python
import csv

# Write data into file
with open(file_name, 'w', newline='') as f:
    
    writer = csv.DictWriter(f, fieldnames=['title', 'lecturer', 'position', 'link'])
    writer.writeheader()

    for video in videos:
        a = video.find('div', class_='vid-title').find('a')

        title = a.get_text()
        link = a.attrs['href']

        try:
            name = video.find('div', attrs={'class':'vid-person'}).get_text()
        except:
            # Add missing value
            name = None

        try:
            position = video.find('div', attrs={'class':'vid-position'}).get_text()
        except:
            position = None

        data = {
            'position': position,
            'link': link,
            'title': title,
            'lecturer': name
        }

        writer.writerow(data)
```
