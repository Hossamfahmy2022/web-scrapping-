# Online Books Data Scrapping

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://books.toscrape.com/catalogue/category/books/mystery_3/index.html"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
articles = soup.find_all('article')
articles
```

```python
len(articles)
```

```python
# Test one article

# Get title
img = articles[0].find('img')
img.attrs['alt']
```

```python
# Get price
p = articles[0].find('div', class_='product_price').find('p', class_='price_color')
p.get_text()
```

```python
# Remove currency and cast to float
float(p.get_text()[2:])
```

```python
# Get price
p = articles[0].find('p', class_='star-rating')
rating = p.attrs['class'][1]
rating
```

```python
rating_map = {
    'One':1,
    'Two':2,
    'Three':3,
    'Four':4,
    'Five':5
}
```

```python
rating_map[rating]
```

```python
# Iterate over all books

for book in articles:
    # Get title
    img = book.find('img')
    title = img.attrs['alt']

    # Get price
    p = book.find('div', class_='product_price').find('p', class_='price_color')
    price = float(p.get_text()[2:])

    # Get rating
    p = book.find('p', class_='star-rating')
    rating = rating_map[p.attrs['class'][1]]

    print(f"Title: {title}, Rating: {rating}, Price: {price}")
    
```
